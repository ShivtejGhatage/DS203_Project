```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import pandas as pd


# mfcc_df = pd.read_csv('MFCC_Meaned_20_with_Song_Names.csv')
mfcc_tsne = pd.read_csv('MFCC_Meaned_20_with_Song_Names.csv')

song_names = mfcc_tsne['Song']  # Save the 'Song' column
mfcc_tsne = mfcc_tsne.drop(columns=['Song']) 

# Apply t-SNE (reduce to 2 dimensions for visualization)
tsne = TSNE(n_components=2, perplexity=45, random_state=42)
mfcc_tsne = tsne.fit_transform(mfcc_tsne)

# Convert the result to a DataFrame for easier handling
mfcc_tsne_df = pd.DataFrame(mfcc_tsne, columns=['Dimension 1', 'Dimension 2'])


# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(mfcc_tsne_df['Dimension 1'], mfcc_tsne_df['Dimension 2'], c='blue', edgecolor='k')
plt.title('t-SNE on 115 Songs MFCCs')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()

mfcc_tsne_df['Song'] = song_names
mfcc_tsne_df.to_csv('tsne.csv')
```
