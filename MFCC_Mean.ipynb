MFCC Combiner here

```python
import os
import pandas as pd

# Path to the folder containing the 115 MFCC CSV files
folder_path = './MFCC_Meaned'  # Update with your folder path

# Initialize an empty list to store data from all songs
mfcc_data_list = []
file_names = []  # List to store the file names

# Loop over all files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        # Read the MFCC file
        file_path = os.path.join(folder_path, file_name)
        mfcc_data = pd.read_csv(file_path, header=None)
        
        # Append the MFCC coefficients as a row in the list
        mfcc_data_list.append(mfcc_data[0].values)
        
        # Store the file name (without .csv) to keep track of which song it is
        file_names.append(file_name.replace('.csv', ''))

# Convert list to DataFrame (115 rows, 20 columns)
mfcc_df = pd.DataFrame(mfcc_data_list)

mfcc_df.to_csv('MFCC_Meaned_20.csv', index=False)


# Add file names as an additional column for tracking
mfcc_df['Song'] = file_names

# Save to a new CSV file with song names
mfcc_df.to_csv('MFCC_Meaned_20_with_Song_Names.csv', index=False)

print(f"Data shape: {mfcc_df.shape}")
```

PCA and tsne

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt



import pandas as pd

mfcc_df = pd.read_csv('MFCC_Meaned_20_with_Song_Names.csv')

song_names = mfcc_df['Song']  # Save the 'Song' column
mfcc_df = mfcc_df.drop(columns=['Song'])  # Use only the MFCC coefficients for PCA and t-SNE
print(mfcc_df.head())

# Step 1: Apply PCA to reduce dimensionality to, say, 10 components


pca = PCA(n_components=10)
mfcc_pca = pca.fit_transform(mfcc_df)


# Step 2: Apply t-SNE on the reduced data from PCA
tsne = TSNE(n_components=2, perplexity=45, random_state=69)
mfcc_tsne_pca = tsne.fit_transform(mfcc_pca)

# Convert t-SNE result to a DataFrame for easy handling
mfcc_tsne_pca_df = pd.DataFrame(mfcc_tsne_pca, columns=['Dimension 1', 'Dimension 2'])


# Plot the results of t-SNE after PCA
plt.figure(figsize=(8, 6))
plt.scatter(mfcc_tsne_pca_df['Dimension 1'], mfcc_tsne_pca_df['Dimension 2'], c='blue', edgecolor='k')
plt.title('t-SNE after PCA on 115 Songs MFCCs')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()

mfcc_tsne_pca_df['Song'] = song_names  # Add the 'Song' column back to the t-SNE results

mfcc_tsne_pca_df.to_csv('MFCCtsnePCA_Mean.csv')
```

```python
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt


import pandas as pd

mfcc_tsne_PCA_df = pd.read_csv('MFCCtsnePCA_Mean.csv')
mfcc_df = pd.read_csv('MFCC_Meaned_20.csv')
print(mfcc_tsne_PCA_df.head())

# Filter the rows where 'Song' equals specific values (1, 2, 3, 4, 5, 6)
song_identifiers = ['mfcc_output_16', 'mfcc_output_48', 'mfcc_output_59', 'mfcc_output_40', 'mfcc_output_73','mfcc_output_3']

filtered_points = mfcc_tsne_PCA_df[mfcc_tsne_PCA_df['Song'].isin(song_identifiers)]


song_names = mfcc_tsne_PCA_df['Song']  # Save the 'Song' column
mfcc_tsne_PCA_df = mfcc_tsne_PCA_df.drop(columns=['Song'])  # Use only the MFCC coefficients for PCA and t-SNE



# Display or assign these points as centroids
initial_centroids = filtered_points[['Dimension 1', 'Dimension 2']].values

print("Initial Centroids:")
print(initial_centroids)

mfcc_tsne_PCA_df = mfcc_tsne_PCA_df.drop(mfcc_tsne_PCA_df.columns[0], axis=1)
print(mfcc_tsne_PCA_df.head())

# Assuming mfcc_df is your DataFrame with 115 songs and 20 MFCC coefficients
gmm = GaussianMixture(n_components=6,means_init=initial_centroids, random_state=42)  # Use 6 components based on your categories
gmm_labels = gmm.fit_predict(mfcc_tsne_PCA_df)


gmm_probabilities = gmm.predict_proba(mfcc_tsne_PCA_df)
probability_df = pd.DataFrame(gmm_probabilities, columns=[f'Cluster_{i}' for i in range(6)])
probability_df['Song'] = song_names 
probability_df.to_csv('Mean_Probability.csv')

# Add GMM labels to the DataFrame for analysis
mfcc_df['GMM Cluster'] = gmm_labels

# Visualizing GMM clusters (you can use t-SNE/PCA results here too)
plt.figure(figsize=(8, 6))
plt.scatter(mfcc_tsne_PCA_df['Dimension 1'], mfcc_tsne_PCA_df['Dimension 2'], c=gmm_labels, cmap='viridis', edgecolor='k')
plt.title('GMM Clusters on t-SNE Reduced Data')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.colorbar(label='Cluster')
plt.show()

mfcc_tsne_PCA_df['GMM Cluster'] = gmm_labels

mfcc_tsne_PCA_df['Song'] = song_names  # Add the 'Song' column back to the t-SNE results
mfcc_tsne_PCA_df.to_csv('Output_GMM_Mean.csv')


# Optional: Print cluster assignment
print(mfcc_df['GMM Cluster'].value_counts())


```
