```python
import pandas as pd

 

for song_no in range(1,10):
    # Load the MFCC file
    file_path = './MFCC-files-v2/' + '0' + str(song_no) + '-MFCC.csv'  # Update with your file path
    mfcc_data = pd.read_csv(file_path, header=None)  # Assuming the file has no header

    # Calculate the mean for each row (axis=1 computes mean across columns for each row)
    mfcc_mean = mfcc_data.mean(axis=1)

    # Convert the result to a DataFrame with 20 rows and 1 column
    mfcc_mean_df = pd.DataFrame(mfcc_mean)

    # Print the result to verify
    print(mfcc_mean_df)

    # Optionally, save the result to a new CSV file
    mfcc_mean_df.to_csv('MFCC_Meaned/mfcc_mean_output' + str(song_no) +'.csv', index=False, header=False)



for song_no in range(10,116):
    # Load the MFCC file
    file_path = './MFCC-files-v2/' + str(song_no) + '-MFCC.csv'  # Update with your file path
    mfcc_data = pd.read_csv(file_path, header=None)  # Assuming the file has no header

    # Calculate the mean for each row (axis=1 computes mean across columns for each row)
    mfcc_mean = mfcc_data.mean(axis=1)

    # Convert the result to a DataFrame with 20 rows and 1 column
    mfcc_mean_df = pd.DataFrame(mfcc_mean)

    # Print the result to verify
    print(mfcc_mean_df)

    # Optionally, save the result to a new CSV file
    mfcc_mean_df.to_csv('MFCC_Meaned/mfcc_mean_output_' + str(song_no) +'.csv', index=False, header=False)
```
